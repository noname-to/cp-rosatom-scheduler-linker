```python

# Таблица 2: Перечень Заявок
ships_df =  pd.read_csv('table_1.csv', sep=';')
ships_df['Дата и время начала плавания'] = pd.to_datetime(ships_df['Дата и время начала плавания'], format='%d.%m.%y %H:%M')
ships_df['Дата и время окончания плавания'] = pd.to_datetime(ships_df['Дата и время окончания плавания'], format='%d.%m.%y %H:%M')
# Таблица 3: Сплоченность в баллах на отрезках маршрута на даты ближайшего месяца
ice_conditions_df = pd.read_csv('path_complexity.csv', sep=';')
ice_conditions_df['Номер ребра'] = pd.Series([i for i in range(1, 15)]) 
ice_conditions_df.drop(['Unnamed: 0'], axis=1, inplace=True)
#ice_conditions = ice_conditions.T
# Таблица 4: Доступность ребра для ледокола
icebreaker_availability_df = pd.read_csv('accessibility.csv', sep=';')
icebreaker_availability_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python

# Множества
ships = ships_df['Наименование'].to_list()
edges = list(range(1, 15))  # 14 рёбер, как указано в geojson
days = list(range(1, 31))  # 30 дней в рассматриваемом месяце
icebreakers = ['50 лет Победы', 'Ямал', 'Таймыр', 'Вайгач']

# Переменные
x = pulp.LpVariable.dicts("ship_route", (ships, edges, days, icebreakers), cat='Binary')

# Дополнительные данные
names = ships_df['Наименование'].to_list()
start_day = {}
end_day = {}
ship_speed = {}
for name in range(len(names)):
    a = ships_df[ships_df['Наименование'] == names[name]]['Дата и время начала плавания'][name].day
    start_day[(names[name], a)] = ships_df[ships_df['Наименование'] == names[name]]['Дата и время начала плавания'][name].day
    end_day[(names[name], a)] = ships_df[ships_df['Наименование'] == names[name]]['Дата и время окончания плавания'][name].day
    ship_speed[(names[name], a)] = ships_df[ships_df['Наименование'] == 'LNG DUBHE']['Скорость, уз'][42]
ice_conditions = {}
for index, row in ice_conditions_df.iterrows():
    edge_id = row['Номер ребра']
    for day in range(1, 31):  # предполагая, что у вас 30 дней
        if edge_id not in ice_conditions:
            ice_conditions[edge_id] = {}
        ice_conditions[edge_id][day] = row[day]
edge_availability = {}
for index, row in icebreaker_availability_df.iterrows():
    edge_id = row['Номер ребра']
    edge_availability[edge_id] = {}
    for icebreaker in ['50 лет Победы', 'Ямал', 'Таймыр', 'Вайгач']:
        edge_availability[edge_id][icebreaker] = row[icebreaker]

```

```python


# Создаем модель линейного программирования
model = pulp.LpProblem("Icebreaker_Optimization", pulp.LpMinimize)

# Определение переменных
# x[i][j][d][k] - 1 если судно i следует по ребру j в день d под проводкой ледокола k, иначе 0
x = {}
for i in ships:
    for j in edges:
        for d in days:
            for k in icebreakers:
                x[i, j, d, k] = pulp.LpVariable(f"x_{i}_{j}_{d}_{k}", 0, 1, pulp.LpBinary)
default_speed = - 10**5
# Целевая функция
default_speed = -10**5
default_date = 1

# Целевая функция
# Целевая функция
model += pulp.lpSum([x[i, j, d, k] * edge_lengths[j] / ship_speed.get((i, start_day.get((i, d), default_speed)), default_speed) 
                     for i in ships for j in edges for d in days for k in icebreakers if start_day.get((i, d), default_speed) is not None]), "Total_Travel_Time"




# Ограничения

# 1. Корабль может следовать по ребру только если это ребро доступно для выбранного ледокола
for i in ships:
    for j in edges:
        for d in days:
            for k in icebreakers:
                model += x[i, j, d, k] <= edge_availability[j][k]

# 2. Корабль может начать движение только после указанной даты начала плавания и до даты окончания
for (ship_name, start_d), val in start_day.items():
    for j in edges:
        for d in days:
            if d < start_d or d > end_day[(ship_name, start_d)]:
                for k in icebreakers:
                    model += x[ship_name, j, d, k] == 0

# 3. Учет условий льда
for i in ships:
    for j in edges:
        for d in days:
            for k in icebreakers:
                start_date_for_ship = start_day.get((i, d), None)
                if start_date_for_ship is not None:
                    # Если сплоченность льда превышает максимальную скорость корабля, корабль не может следовать по этому ребру
                    model += x[i, j, d, k] * ice_conditions[j][d] <= ship_speed.get((i, start_date_for_ship), default_speed)


# 4. Формирование караванов: ледокол может сопровождать максимум 3 корабля на одном ребре
for k in icebreakers:
    for j in edges:
        model += pulp.lpSum([x[i, j, d, k] for i in ships for d in days]) <= 3

# Решаем задачу
model.solve()

# Выводим результаты
for i in ships:
    for j in edges:
        for d in days:
            for k in icebreakers:
                if x[i, j, d, k].varValue > 0.5:  # Если переменная равна 1
                    print(f"Ship {i} follows edge {j} on day {d} under the guidance of icebreaker {k}")

```

```python
status = model.status

if status == pulp.LpStatusOptimal:
    print("Оптимальное решение найдено.")
elif status == pulp.LpStatusInfeasible:
    print("Модель неразрешима. Нет допустимого решения.")
elif status == pulp.LpStatusUnbounded:
    print("Модель неограничена.")
elif status == pulp.LpStatusUndefined:
    print("Статус решения не определен.")
else:
    print("Неизвестный статус решения.")
/
```
